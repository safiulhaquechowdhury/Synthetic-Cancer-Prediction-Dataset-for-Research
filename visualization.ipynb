# Heatmap

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Generate the correlation matrix
correlation_matrix = dataset.corr()

# Set up the matplotlib figure
plt.figure(figsize=(10, 8))

# Generate a heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='Blues', fmt=".2f", linewidths=0.5)

# Set the title and adjust layout
plt.title('Correlation Heatmap OF Cancer Prediction')
plt.tight_layout()

# Show the plot
plt.show()
```

# Density Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'dataset' is your DataFrame

# Create subplots
fig, axes = plt.subplots(2, 3, figsize=(15, 10))

# Define the columns for the density plots
columns = ['Gender','Age','Smoking','Fatigue','Allergy','Cancer']
colors = ['blue', 'green', 'red', 'purple', 'orange', 'brown']
# Flatten the axes array for easy iteration
axes = axes.flatten()

# Generate density plots for each column
for i, col in enumerate(columns):
    sns.kdeplot(data=dataset[col], shade=True, ax=axes[i])
    axes[i].set_title(f'Density Plot for {col}')
    axes[i].set_xlabel('Value')
    axes[i].set_ylabel('Density')

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```

# Histogram

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'dataset' is your DataFrame

# Create subplots
fig, axes = plt.subplots(2, 3, figsize=(15, 10))

# Define the columns for the histograms
columns = ['Gender','Age','Smoking','Fatigue','Allergy','Cancer']
colors = ['blue', 'green', 'red', 'purple', 'orange', 'brown']
# Flatten the axes array for easy iteration
axes = axes.flatten()

# Define colors
colors = ['red', 'blue', 'green', 'purple', 'orange', 'yellow']

# Generate histograms for each column with specified colors
for i, (col, color) in enumerate(zip(columns, colors)):
    sns.histplot(data=dataset[col], color=color, ax=axes[i], kde=True)  # Use sns.histplot for histograms with KDE
    axes[i].set_title(f'Histogram for {col}')
    axes[i].set_xlabel('Value')
    axes[i].set_ylabel('Frequency')

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```

# Box Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'dataset' is your DataFrame

# Create subplots
fig, axes = plt.subplots(2, 3, figsize=(15, 10))

# Define the columns for the box plots
columns = ['Gender','Age','Smoking','Fatigue','Allergy','Cancer']

# Flatten the axes array for easy iteration
axes = axes.flatten()

# Define colors
colors = ['red', 'blue', 'green', 'purple', 'orange', 'yellow']

# Generate box plots for each column with specified colors
for i, (col, color) in enumerate(zip(columns, colors)):
    sns.boxplot(x=dataset[col], color=color, ax=axes[i])
    axes[i].set_title(f'Box Plot for {col}')
    axes[i].set_xlabel('Value')

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```
